## Question 9
If you concatenate the first three digits of the most common ‘Max_temperature_C’ with the last three letters of the least common 'Crop_type', what string do you get?
Note: Use the first mode if there are multiple modes

```python
most_common_max_temp = df['Max_temperature_C'].mode()[0]  # Get the first mode if there are multiple modes
first_three_digits = str(most_common_max_temp)[:3]  # Take the first 3 digits

# Step 2: Find the least common 'Crop_type'
least_common_crop_type = df['Crop_type'].value_counts().idxmin()  # Find the least common crop type
last_three_letters = least_common_crop_type[-3:]  # Take the last 3 letters

# Step 3: Concatenate the results
result_string = first_three_digits + last_three_letters

print(f"The resulting string is: {result_string}")
```

## Question 10
Write Python code to create a violin plot visualising the distribution of 'Annual_yield' across different 'Elevation' ranges. Use the provided elevation range categories (Low: < 300m, Medium: 300m - 600m, High: > 600m) to categorise the data before plotting.  Examine the violin plot displaying the distribution of 'Annual Yield' across three elevation categories (Low, Medium, High).

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Categorize the 'Elevation' column into Low, Medium, High
bins = [0, 300, 600, float('inf')]  # Define the elevation ranges
labels = ['Low', 'Medium', 'High']  # Labels for each range
df['Elevation_category'] = pd.cut(df['Elevation'], bins=bins, labels=labels, right=False)

# Step 2: Create a violin plot for 'Annual_yield' across 'Elevation_category'
plt.figure(figsize=(10, 6))  # Set the size of the plot
sns.violinplot(x='Elevation_category', y='Annual_yield', data=df, inner="quart", palette="Set2")

# Step 3: Customize the plot (title, labels, etc.)
plt.title('Distribution of Annual Yield Across Different Elevation Categories', fontsize=16)
plt.xlabel('Elevation Category', fontsize=14)
plt.ylabel('Annual Yield', fontsize=14)

# Show the plot
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

## Question 11
Assuming each 'Crop_type' contributes an integer value equal to its length (e.g., 'wheat' contributes 5), write a recursive function to sum the integer values for each unique crop type in the dataset. What is the sum?

```python
# Step 1: Get unique crop types
unique_crop_types = df['Crop_type'].unique()

# Step 2: Define the recursive function
def sum_crop_values(crop_list):
    # Base case: If the list is empty (length 0), return 0
    if len(crop_list) == 0:
        return 0
    # Recursive case: Get the length of the first crop type and sum with the rest
    return len(crop_list[0]) + sum_crop_values(crop_list[1:])

# Step 3: Calculate the total sum for all unique crop types
total_sum = sum_crop_values(unique_crop_types)

print(f"The total sum of crop values is: {total_sum}")
```

## Question 12
Write Python code to perform a t-test comparing the average 'Annual_yield' between 'coffee' and 'banana' crop types using scipy.stats. What is the p-value, rounded to three decimal places?

```python
from scipy import stats


# Step 1: Filter data for 'coffee' and 'banana' crop types
coffee_yield = df[df['Crop_type'] == 'coffee']['Annual_yield']
banana_yield = df[df['Crop_type'] == 'banana']['Annual_yield']

# Step 2: Perform t-test
t_stat, p_value = stats.ttest_ind(coffee_yield, banana_yield, nan_policy='omit')  # nan_policy='omit' to ignore NaNs

# Step 3: Print the p-value rounded to three decimal places
print(f"The p-value of the t-test is: {p_value:.3f}")
```
